```python
text_cols = ['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior']
for col in text_cols:
    df[col] = df[col].fillna('unknown').str.lower().str.strip()
```

```python
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce')
most_common_date = df['saledate'].mode()[0]
df['saledate'] = df['saledate'].fillna(most_common_date)
```

```python
for col in ['condition', 'odometer', 'mmr', 'sellingprice']:
    df[col] = df[col].fillna(df[col].median()).astype(int)
```

```python
for col in ['year','odometer','mmr','sellingprice','condition']:
    df[col]=df[col].astype(int)
```

```python
#  Standardize text values
for col in ['color','interior','seller']:
    df[col]=df[col].str.lower().str.strip()
```

```python
df['state'] = df['state'].str.upper().str.strip()
```
